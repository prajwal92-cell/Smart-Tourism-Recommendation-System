```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics.pairwise import cosine_similarity
```

```python
city = pd.read_excel("City.xlsx")
continent = pd.read_excel("Continent.xlsx")
country = pd.read_excel("Country (1).xlsx")
item = pd.read_excel("Item.xlsx")
mode = pd.read_excel("Mode.xlsx")
region = pd.read_excel("Region.xlsx")
transaction = pd.read_excel("Transaction.xlsx")
type_df = pd.read_excel("Type.xlsx")
user = pd.read_excel("User.xlsx")
```

```python


df = transaction.merge(user, on="UserId", how="left")

df = df.merge(item, on="AttractionId", how="left")

df = df.merge(city, left_on="AttractionCityId", right_on="CityId", how="left")

df = df.merge(country, left_on="CountryId_y", right_on="CountryId", how="left")

df = df.merge(region, left_on="RegionId_y", right_on="RegionId", how="left")

df = df.merge(continent, left_on="ContinentId_y", right_on="ContinentId", how="left")

df = df.merge(type_df, on="AttractionTypeId", how="left")
```

```python
print(df.shape)
df.head()
```

```python
# Remove unwanted duplicate columns
columns_to_drop = [col for col in df.columns if col.endswith('_x')]
df.drop(columns=columns_to_drop, inplace=True)

# Rename _y columns properly
df.columns = df.columns.str.replace('_y', '', regex=False)

print("Final Columns:")
print(df.columns)
print("Final Shape:", df.shape)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Keep only valid ratings
df = df[df["Rating"].between(1, 5)]

# Fill missing values
df.ffill(inplace=True)

df.info()
```

```python
#visit mode distribution
plt.figure()
sns.countplot(x="VisitMode", data=df)
plt.xticks(rotation=45)
plt.title("Visit Mode Distribution")
plt.show()
```

```python
# encode categorical columns 
le = LabelEncoder()

categorical_cols = [
    "Continent",
    "Country",
    "Region",
    "CityName",
    "AttractionType",
    "VisitMode"
]

for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
#feature selection 
features = [
    "Continent",
    "Country",
    "Region",
    "CityName",
    "VisitYear",
    "VisitMonth",
    "AttractionType"
]

X = df[features]

y_reg = df["Rating"]
y_clf = df["VisitMode"]
```

```python
# train - test split 
X_train, X_test, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

_, _, y_train_clf, y_test_clf = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)
```

```python
# regression model (rating prediction)
reg_model = LinearRegression()
reg_model.fit(X_train, y_train_reg)

y_pred_reg = reg_model.predict(X_test)

print("R2 Score:", r2_score(y_test_reg, y_pred_reg))
print("RMSE:", np.sqrt(mean_squared_error(y_test_reg, y_pred_reg)))
```

```python
# classification model
clf_model = RandomForestClassifier()
clf_model.fit(X_train, y_train_clf)

y_pred_clf = clf_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print(classification_report(y_test_clf, y_pred_clf))
```

```python
# create user-item matrix 
user_item_matrix = df.pivot_table(
    index="UserId",
    columns="AttractionId",
    values="Rating"
).fillna(0)
```

```python
#compute similarity
user_similarity = cosine_similarity(user_item_matrix)

similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)
```

```python
# Recommendation function 
def recommend_attractions(user_id, top_n=5):
    
    similar_users = similarity_df[user_id].sort_values(ascending=False)[1:6]
    similar_users_ids = similar_users.index
    
    recommended = user_item_matrix.loc[similar_users_ids].mean().sort_values(ascending=False)
    
    visited = user_item_matrix.loc[user_id]
    recommended = recommended[visited == 0]
    
    return recommended.head(top_n)
```

```python
# seasonal travel intelligence 

def get_season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Summer"
    elif month in [6,7,8]:
        return "Monsoon"
    else:
        return "Autumn"

df["Season"] = df["VisitMonth"].apply(get_season)
df["Season"].head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select only numeric features for clustering
cluster_features = df[[
    "Rating",
    "SatisfactionScore"
]]

# Scale the data (IMPORTANT for KMeans)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(cluster_features)

# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
df["Cluster"] = kmeans.fit_predict(scaled_features)

print("Clustering completed successfully!")
```
